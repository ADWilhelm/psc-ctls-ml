# How many cells are currently identified?
For how many cells are there dictionary entries which identify both CTLs?

```python
import pickle

# load dictionary

with open('pickles/dict_all_cells_v2.pkl', 'rb') as f:
    name_to_CID_dict = pickle.load(f)
    f.close()
```

```python
# add empty entry for cells without an etl or htl
print(len(name_to_CID_dict))
name_to_CID_dict['none'] = None
print(len(name_to_CID_dict))
```

```python
# load dataset and transform entries where multiple materials are separated 
# by a ";" into lists

with open('pickles/df_all_ctls.pkl', 'rb') as f:
    df_all_ctls = pickle.load(f)

for index, _ in df_all_ctls.iterrows():
    i = 0
    while i < len(df_all_ctls.at[index, "etl"]):
        if ";" in df_all_ctls.at[index, "etl"][i]:
            elements = df_all_ctls.at[index, "etl"][i].split(";")
            df_all_ctls.at[index, "etl"] = df_all_ctls.at[index, "etl"][:i] + elements + df_all_ctls.at[index, "etl"][i+1:]
        i += 1
```

```python
import pandas as pd
# Create a DataFrame from the dictionary
df_new = pd.DataFrame(list(name_to_CID_dict.items()), columns=['Material', 'Value'])

# Create the 'etl' and 'htl' columns (about 3 minutes)
df_new['etl'] = df_new['Material'].apply(lambda x: df_all_ctls[df_all_ctls['etl'].apply(lambda y: x in y)].index.tolist())
df_new['htl'] = df_new['Material'].apply(lambda x: df_all_ctls[df_all_ctls['htl'].apply(lambda y: x in y)].index.tolist())
```

```python
# ETLs
none_count_etl = 0
etl_identified = []
etl_count = 0

for index, row in df_all_ctls.iterrows():
    print(f"Index: {index}/43119")
    num_etl_materials = len(row['etl'])
    current_num = 0
    for material in row['etl']:
        if material != "none":
           if pd.isna(df_new[df_new['Material'] == material]['Value'].values[0]):
               break
           else:
               current_num = current_num + 1
        else:
            none_count_etl = none_count_etl + 1


    if current_num == num_etl_materials:
        etl_count = etl_count + 1
        etl_identified.append(True)
    else:
        etl_identified.append(False)
```

```python
# resulting count of entries with an identified etl
print(f"{etl_count} entries have an identified ETL.")
print(f"{none_count_etl} entries use no ETL material.")
```

```python
# HTLs (these often contain "none" as a material, so the code here also counts
# those cells)
none_count_htl = 0
htl_count = 0
htl_identified = []

for index, row in df_all_ctls.iterrows():
    print(f"Index: {index}/43119")
    num_etl_materials = len(row['htl'])
    current_num = 0
    for material in row['htl']:
        if material != "none":
           if pd.isna(df_new[df_new['Material'] == material]['Value'].values[0]):
               break
           else:
               current_num = current_num + 1
        else:
            none_count_htl = none_count_htl + 1


    if current_num == num_etl_materials:
        htl_count = htl_count + 1
        htl_identified.append(True)
    else:
        htl_identified.append(False)
```

```python
# resulting count of entries with an identified HTL
print(f"{htl_count} entries have an identified HTL.")
print(f"{none_count_htl} entries use no HTL material.")
```

```python
# add identification to materials dataframe
df_all_ctls['etls_identified'] = etl_identified
df_all_ctls['htls_identified'] = htl_identified

# where etls or htls are none, change them to identified = True
df_all_ctls.loc[df_all_ctls['etl'].apply(lambda x: x == ['none']), 'etls_identified'] = True
df_all_ctls.loc[df_all_ctls['htl'].apply(lambda x: x == ['none']), 'htls_identified'] = True

# create 'both_identified' column using updated 'etls_identified' and 'htls_identified' columns
df_all_ctls['both_identified'] = df_all_ctls['etls_identified'] & df_all_ctls['htls_identified']
```

```python
def check_both_none(row):
    return (row['etl'][0] == 'none') and (row['htl'][0] == 'none')

none_count_both = df_all_ctls.apply(check_both_none, axis=1).sum()
print(f"{none_count_both} entries use no ETL or HTL material.")
```

```python
# both etl and htl is identified for this many cells:
# note that the entries without htl are not included, so the number of entries
# that are sufficiently identified is up to ~2800 higher.
df_all_ctls['both_identified'].sum()
```

# How often were identified materials used?

How often does an identified material appear in the db?

```python
sum_etl = 0
sum_htl = 0

for index, material in enumerate(df_new['Material']):
    if not pd.isna(df_new[df_new['Material'] == material]['Value'].values[0]):
        current_etl_sum = len(df_new[df_new['Material'] == material]['etl'].values[0])
        current_htl_sum = len(df_new[df_new['Material'] == material]['htl'].values[0])
        print(f"Etls with {material}: {current_etl_sum}")
        print(f"Htls with {material}: {current_htl_sum}")
        print(f"{index+1}/2591")
        sum_etl = sum_etl + current_etl_sum
        sum_htl = sum_htl + current_htl_sum
```

```python
print(f"Etls identified: {sum_etl}")
print(f"Htls identified: {sum_htl}")
```

# Unidentified materials: How many cells pertain to each?

This is interesting to see if there are unidentified materials which are used in many cells. In such cases it might be worth the time to identify them manually.

```python
unidentified_materials = []
no_cells = []

for index, row in df_new.iterrows():
    if pd.isna(row['Value']):
        unidentified_materials.append(row['Material'])
        no_cells.append(len(row['etl']) + len(row['htl']))

df_unid = pd.DataFrame({'Material': unidentified_materials, 'Cell_amount': no_cells})
```

```python
df_unid.sort_values(by = 'Cell_amount', ascending=False).to_excel('unidentified.xlsx', sheet_name='Unidentified')
```
