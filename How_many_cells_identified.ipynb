```python
import pandas as pd
# Create a DataFrame from the dictionary
df_new = pd.DataFrame(list(name_to_CID_dict.items()), columns=['Material', 'Value'])

# Create the 'etl' and 'htl' columns (about 3 minutes)
df_new['etl'] = df_new['Material'].apply(lambda x: df_all_ctls[df_all_ctls['etl'].apply(lambda y: x in y)].index.tolist())
df_new['htl'] = df_new['Material'].apply(lambda x: df_all_ctls[df_all_ctls['htl'].apply(lambda y: x in y)].index.tolist())
```

```python
df_new.head(20)

sum_etl = 0
sum_htl = 0

for index, material in enumerate(df_new['Material']):
    if not pd.isna(df_new[df_new['material' == material]]['Value']):
        sum_etl = sum_etl + len(df_new[df_new['material' == material]]['etl'])
        sum_htl = sum_htl + len(df_new[df_new['material' == material]]['htl'])
    if index > 20:
        break
```
